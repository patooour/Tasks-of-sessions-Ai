```python
df1 = pd.DataFrame({
    'cities':['mahalla','tanta','mansora','portsaid'],
    'temp':[10,20,30,40]
})
```

```python
df2 = pd.DataFrame({
    'cities':['mahalla','tanta','mansora','portsaid'],
    'age':[2,1,5,9]
})
```

```python
df3 = pd.DataFrame({
    'cities':['mahalla','tanta','mansora','portsaid'],
    'population':[5000,1000,80000,100000],
    
})
```

```python
# first way

frames = [df1,df2,df3]
www = pd.concat(frames,ignore_index=True)
www
```

```python
# second way

qqq = pd.merge(pd.merge(df1,df2,on="cities"),df3 ,on="cities")
qqq
```

```python
# third way
from functools import reduce

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['cities'],how='outer'), frames)
df_merged
```
